```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#line plotting
x = np.array([2,8])
y = np.array([3,10])
plt.plot(x, y)
plt.show()
```

```python
#pointer plotting
x = np.array([2,8])
y = np.array([3,10])
plt.plot(x, y,'o')
plt.show()
```

```python
#making a Zig-Zag line
x = np.array([1,2,3,5,7])
y = np.array([6,4,10,3,8])

plt.plot(x, y)
plt.show()
```

```python
#Default X-points => 0,1,2,3,4...similar to len(y)
ypoints = np.array([2,4,6,3])
plt.plot(ypoints)
plt.show()
```

## Markers
Marker	Description
'o'=>	Circle	
'*'=>	Star	
'.'=>	Point	
','=>	Pixel	
'x'=>	X	
'X'=>	X (filled)	
'+'=>	Plus	
'P'=>	Plus (filled)	
's'=>	Square	
'D'=>	Diamond	
'd'=>	Diamond (thin)	
'p'=>	Pentagon	
'H'=>	Hexagon	
'h'=>	Hexagon	
'v'=>	Triangle Down	
'^'=>	Triangle Up	
'<'=>	Triangle Left	
'>'=>	Triangle Right	
'1'=>	Tri Down	
'2'=>	Tri Up	
'3'=>	Tri Left	
'4'=>	Tri Right	
'|'=>	Vline	
'_'=>	Hline

```python
plt.plot(x,y, marker='P')
plt.show()
```

## Format Strings fmt
You can use also use the shortcut string notation parameter to specify the marker.

This parameter is also called fmt, and is written with this syntax:

`marker|line|color` 

```python
plt.plot(x, y, 'o:r')
```

```python
#size of marker
plt.plot(x, y, marker='^',ms=15)
```

```python
# marker edge color and marker face color
plt.plot(x, y, marker = 'P',ms=15, mec='r', mfc='y')
```

## Labels & Titles

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot')
plt.show()
```

```python
#font size
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

```python
#Title Position : Legal values are: 'left', 'right', and 'center'. Default value is 'center'
plt.plot(x, y)
plt.title("Plot",loc='left',fontdict = font1)
plt.show()
```

```python
#grid
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
#choose which axis to show grid
plt.plot(x, y)
plt.grid(axis='x')
plt.show()
```

```python
plt.plot(x, y)
plt.grid(axis='y')
plt.show()
```

```python
#line properties of grid
plt.plot(x, y)
plt.grid(linewidth = 0.5, color='r', linestyle='--')
plt.show()
```

# Scatter Plot

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.xlabel('Car age')
plt.ylabel('Speed')
plt.show()
```

```python
#day one, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, c='hotpink')

#day two, the age and speed of 15 cars:
x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, c='b')

plt.xlabel('Car age')
plt.ylabel('Speed')
plt.show()
```

```python
#color each dot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

```python
#using built-in color map => viridis
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])

plt.scatter(x, y, c=colors, cmap='viridis')

plt.show()
```

```python
#sizes
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes)

plt.show()
```

```python
#Transparency
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5)

plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
size = 10*np.random.randint(100, size=(100))
color = np.random.randint(100, size=(100))

plt.scatter(x, y, size, color, alpha=0.5, cmap='nipy_spectral')
plt.colorbar()
plt.show()
```

# Bar Graph

```python
#Horizontal Bars
x = np.array([21,23,19,20])
y = np.array([5,3,2,8])
plt.barh(x, y)
plt.show()
```

```python
#bar width
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([5,3,2,8])
plt.bar(x, y, width = 0.1)  #default = 0.8
plt.show()
```

# Pie Chart

```python
y = np.array([25,15,10,40,10])
plt.pie(y)
plt.show()
print('By default the plotting of the first wedge starts from the x-axis and move counterclockwise:')
```

```python
#adding label
y = np.array([25,15,10,40,10])
plt.pie(y, labels=['Apple', 'Banana','Cherries','Dates','Mango'])
plt.show()
```

```python
#changing start angle, which is initially set to x axis
y = np.array([25,15,10,40,10])
plt.pie(y, labels=['Apple', 'Banana','Cherries','Dates','Mango'], startangle=90)
plt.show()
```

```python
#explode 
y = np.array([25,15,10,40,10])
plt.pie(y, labels=['Apple', 'Banana','Cherries','Dates','Mango'], explode=[0.2,0,0,0,0])
plt.show()
```

```python
#shadow
y = np.array([25,15,10,40,10])
plt.pie(y, labels=['Apple', 'Banana','Cherries','Dates','Mango'], shadow=True, explode=[0.2,0,0,0,0])
plt.show()
```

```python
#color
y = np.array([25,15,10,40,10])
plt.pie(y, labels=['Apple', 'Banana','Cherries','Dates','Mango'], shadow=True, explode=[0.2,0,0,0,0], 
        colors=["black", "hotpink", "b", "#4CAF50",'red'])
plt.show()
```

'r' - Red
'g' - Green
'b' - Blue
'c' - Cyan
'm' - Magenta
'y' - Yellow
'k' - Black
'w' - White

```python
# Legend
y = np.array([25,15,10,40,10])
plt.pie(y, labels=['Apple', 'Banana','Cherries','Dates','Mango'], shadow=True, explode=[0.2,0,0,0,0])
plt.legend()
plt.show()
```

```python
# Giving title to Legend
y = np.array([25,15,10,40,10])
plt.pie(y, labels=['Apple', 'Banana','Cherries','Dates','Mango'], shadow=True, explode=[0.2,0,0,0,0])
plt.legend(title='Fruits')
plt.show()
```
