```python
import seaborn as sns
import pandas as pd
```

# Distribution Plot

**Used to deal with numerical values**
* distplot
* pairplot
* joinplot

```python
df = pd.read_csv('Automobile_data.csv')
del df['symboling']
del df['wheel-base']
df.head()
```

## Correlation with Heatmap

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## Joinplot

```python
# It is used for bivariate analysis
sns.jointplot(x='height', y='width', data=df, kind='hex')
```

```python
sns.jointplot(x='height', y='width', data=df, kind='reg')
```

# Pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='drive-wheels')
```

## Dist plot

```python
sns.distplot(df['length'])
```

# Categorical Plots

**deal with categorical values**
* boxplot
* violinplot
* countplot
* barplot

## Count plot

```python
sns.countplot('body-style', data=df)
```

```python
sns.countplot(y='body-style', data=df)
```

```python
sns.countplot('drive-wheels', data=df, hue ='body-style')
```

## Bar plot

### It will show categorical features comparing with numerical features(Any), unlike Countplot shows frequency of one feature.

```python
sns.barplot(x='body-style', y='height', data=df)
```

## Box plot
Here, starting shows 0 percentile and then increases by 25 percentile by a line.Also the dots in the figure shows outlier.
Comparison can be with categorical data or numerical data.

```python
sns.boxplot('body-style', 'height', data=df)
```

```python
sns.boxplot('fuel-type','engine-size', data=df, hue='body-style', palette='rainbow')
```

```python
sns.boxplot(data=df)
```

## Violin plot
It shows density in form of box plot

```python
sns.violinplot('drive-wheels', 'height',data=df, hue='fuel-type')
```
